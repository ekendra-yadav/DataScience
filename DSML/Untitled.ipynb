## Recursive Function (Recursion)

````python
#Fibonacci 
def fib(n):
    if n==0:
        return 0
    elif n== 1:
        return 1
    # recursive case
    else:
        return fib(n-1) + fib(n-2)
````

````python
%%time 
fib(100)
````

## Class and object 

Calculating Student Results
Develop a class to accept a student's name and marks in three subjects, then calculate and display the total and average marks.
 

````python
class StudentResult:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def total(self):
        return sum(self.marks)

    def avg(self):
        return self.total() / len(self.marks)

    def display(self):
        print(f"Name:{self.name}")
        print("-" * 50)
        print(f"Total marks obtained: {self.total()}")
        print(f"Average marks obtained: {self.avg()}")
````

````python
ekendra = StudentResult("Ekendra",[90,85,98])
````

````python
ekendra.display()
````

### Inheritance 

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

class Manager(Employee):
    def __init__(self, name, age, employee_id, department="not_assigned"):
        super().__init__(name, age, employee_id)
        self.department = department
    
    def get_details(self):
        print(f"{self.name}, {self.age}, {self.employee_id}, {self.department}")
````

````python
m = Manager("Ram", 30, "E101", "HR")
m.get_details()
````

````python
class Point:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def __str__(self):
        return f"Point({self.a},{self.b})"

    def __add__(self,other):
        a = self.a + other.a
        b = self.b + other.b
        return Point(a,b)

    def __mul__(self,other):
        a = self.a * other.a
        b = self.b * other.b
        return Point(a,b)

A = Point(3,4)
B = Point(6,7)
````
