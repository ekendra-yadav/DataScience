## List Slicing 

```python
numbers = list(range(1,101))
print(numbers)
```

```python
odd_numbers = numbers[0::2]
print(odd_numbers)
```

```python
even_numbers = numbers[1::2]
print(even_numbers)
```

```python
numbers.clear()   #To clear the list used for (memory kahli garna 
numbers
```

### copy by value or by memory reference

```python
numbers = [1,2,3]
numbers_copy=numbers
numbers_copy.extend([4,5,6])
print(numbers_copy)
print(numbers)
```

```python
numbers = [1,2,3]
numbers_copy=numbers.copy()
numbers_copy.extend([4,5,6])
print(numbers_copy)
print(numbers)
```

# Nested Loops


```python
adj = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]

for fruit in fruits:
    for ad in adj:
        print(ad,fruit)
```

```python
n = int(input("Enter the number: "))
for i in range(1,11):
    print(f"{n} x {i} = {n*i}")
```

### Multiple arguments in loop

```python
person = {
    "name": "Ekendra",
    "age" : 20,
    "gender" : "male"
}

person
```

```python
for p in person:
    print(p)
    
```

```python
for p in person.items():
    print(p)
```

```python
for (key,value) in person.items():
    print(key)
```

```python
for (key,value) in person.items():
    print(value)
```

```python
coord = [[1,23],[2,34],[3,45]]

for cor in coord:
    print(cor)
```

```python
coord = [[1,23],[2,34],[3,45]]

for x,y in coord:
    print(x,y)
```
