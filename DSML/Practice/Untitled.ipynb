## Constructor
In Python, a constructor is a special type of method used to initialize the object of a Class. The constructor will be executed automatically when the object is created. If we create three objects, the constructor is called three times and initialize each object.

The main purpose of the constructor is to declare and initialize instance variables. It can take at least one argument that is self. The `__init()__` method is called the constructor in Python. In other words, the name of the constructor should be `__init__(self)`.

A constructor is optional, and if we do not provide any constructor, then Python provides the default constructor. Every class in Python has a constructor, but it's not required to define it.

```python
class Student:
    '''
    # default constructor
    def __init__(self):
        pass
    '''
    #parameterized constructor
    def __init__(self,fullname,marks):
        self.name = fullname
        self.marks = marks
        print("adding new student in database...")

s1 = Student("Ekendra Yadav",{"Science" : 92,
                        "Mathematics" : 98,
                        "English" : 97,
                        "Computer" : 100,
                        "Nepali" : 96})
print(f"The name of the student is {s1.name} and his marks are {s1.marks}.")
```

## Class and Object(Instance) Attributes

In Python, attributes are variables that store data within a class or an object. There are two types: class attributes, which are shared by all instances of a class, and instance attributes, which are unique to each individual object. 

```python
class Dog:
    species = "canine"  # Class attribute

    def __init__(self, name):
        self.name = name # Instance attribute
```

```python
# Creating objects (instances) from the Dog class
my_dog = Dog("Buddy")
your_dog = Dog("Lucy")

# Accessing instance attributes (unique to each)
print(f"My dog's name: {my_dog.name}")   # Output: My dog's name: Buddy
print(f"Your dog's name: {your_dog.name}") # Output: Your dog's name: Lucy

# Accessing class attribute (shared)
print(f"My dog is a {my_dog.species}")   # Output: My dog is a canine
print(f"Your dog is a {your_dog.species}") # Output: Your dog is a canine
```

## Methods

Methods are the function that belongs to object 

```python
class Student:
    def __init__(self, name):
        self.name = name

    def welcome(self):
        print("Welcome",self.name)  # Method

s1 = Student("Ekendra")
s1.welcome()  # calling a Method 
    
```

#### Create Student class that takes names and marks of 3 subject as arguiments in constructor then create a method to print the average.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def avg(self):
        return sum(self.marks)/len(self.marks)

s1 = Student("Ekendra",[99,98,97])
print(f"Hello {s1.name} your marks is {s1.marks}")
print(f"Your average marks is {s1.avg()}")
```
